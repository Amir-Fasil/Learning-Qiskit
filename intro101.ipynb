```python
import qiskit
from qiskit_ibm_runtime import QiskitRuntimeService
from qiskit.circuit import QuantumRegister, ClassicalRegister, QuantumCircuit 
import matplotlib.pyplot as plt
```

### The Deutch Algorithm using Compositing method for adding gates

```python
# Defining registers that store qubits and classical bits for different purposes
quantum_reg = QuantumRegister(2, name  = "data")
classical_reg = ClassicalRegister(2, name = 'result')

circuit = QuantumCircuit(quantum_reg, classical_reg)
circuit.x(quantum_reg[1])
circuit.h(quantum_reg[1])
circuit.draw("mpl")
```

```python
circuit_b = QuantumCircuit(quantum_reg, classical_reg)
circuit_b.cx(quantum_reg[0], quantum_reg[1])
circuit_b.draw("mpl")
```

```python
circuit_c = QuantumCircuit(quantum_reg, classical_reg)
circuit_c.h(quantum_reg[0])
circuit_c.measure(quantum_reg[0], classical_reg[0])
circuit_c.draw("mpl")
```

```python
circuit.compose(circuit_b, qubits = [0,1], clbits= [0,1], inplace = True)
circuit.compose(circuit_c, qubits = [0,1], clbits= [0,1], inplace = True)
circuit.draw("mpl")
```
