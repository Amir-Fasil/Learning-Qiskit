```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### Line Graphs

```python
x = np.arange(10)
y = np.square(x)
z = np.power(x,3)
e = np.exp(x)

fig , ax = plt.subplots()
ax.plot(x,y, label='y = x^2', marker = 'o')
ax.plot(x,z, label='y = x^3', marker = 's')
ax.plot(x,e, label='y = e^x', marker = '^')
ax.legend()
ax.set_xlabel("x_axis")
ax.set_ylabel("y_axis")
ax.grid(True, alpha = 0.51)
plt.show()
```

### Scatters

```python
x = np.random.randn(100)
y = 2 * np.random.randn(100) + 3
size = 10 * np.random.randn(100) + 50
color = np.random.randn(100)

fig, ax =  plt.subplots()
ax.scatter(x,y, marker='o', alpha=0.4, s = size, c = color, cmap='viridis')
ax.set_title("Scatter Plot Example")
ax.set_xlabel("X values")
ax.set_ylabel("Y values")
plt.show()
```

### Bar Charts

```python
people = np.array(["Amir", "Eba", "Sema", "Pen", "Wale" , "Nahom"])
height = 10 * np.random.randn(len(people)) + 175
age = 5 * np.random.randn(len(people)) + 25

# print(len(people), len (height), len(age))
lab_loc = np.arange(len(people)) # [0,1,2,3,4,5]
print(lab_loc)
width = 0.35
print(lab_loc - width/2)
print(lab_loc + width/2)

fig , ax = plt.subplots()
prop1 = ax.bar(lab_loc - width/2, height = height, width = 0.35, color = 'skyblue', edgecolor='black', alpha=0.7)
prop2 = ax.bar(lab_loc + width/2, height = age, width = 0.35, color = 'lightgreen', edgecolor='black', alpha=0.7)

ax.set_xticks(lab_loc)
ax.set_xticklabels(people)
ax.set_title("Height and Age of People")


ax.set_xlabel("People")
ax.set_ylabel("Height")
plt.show()
```
