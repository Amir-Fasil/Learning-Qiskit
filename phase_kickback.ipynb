```python
import qiskit
import numpy as np
from qiskit.quantum_info import Statevector, Operator, partial_trace, DensityMatrix
from qiskit.circuit import QuantumCircuit, QuantumRegister, ClassicalRegister
import matplotlib.pyplot as plt
```

```python
def phase_kickback():
        
        """The below circuit is implemented to showcase teh concept of pashekickback by using the deutch algorithm.
        In this circuit, we will see how the phase of the qubit is changed based on the function implemented by the oracle.
        """

        # Defining a register
        quantum_reg = QuantumRegister(2, 'qreg')
        classical_reg = ClassicalRegister(2, 'creg')

        # Creating a Qunatum Circuit
        circuit = QuantumCircuit(quantum_reg, classical_reg)
        circuit.x(quantum_reg[1])
        circuit.h(quantum_reg[1])   
        circuit.cx(quantum_reg[0], quantum_reg[1])
        circuit.h(quantum_reg[0])

        return circuit
```
